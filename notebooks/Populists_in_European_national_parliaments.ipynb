## 1. Gather national election results

The data comes from [ParlGov](http://www.parlgov.org/), a database with information on parties, elections and cabinets worldwide. We need the results from all national parliamentary elections in the EU28 since 1998. (Please note that the partys' short names used by ParlGov can differ from the usual abbreviations, sometimes.)

```python
url = 'http://www.parlgov.org/static/data/development-utf-8/view_election.csv'

selection = {
    'election_type': ['parliament'],
    'country_name_short': ['AUT', 'BEL', 'BGR', 'CYP', 'CZE', 'DEU', 'DNK', 'ESP', 'EST', 'FIN',
                           'FRA', 'GBR', 'GRC', 'HRV', 'HUN', 'IRL', 'ITA', 'LTU', 'LUX', 'LVA',
                           'MLT', 'NLD', 'POL', 'PRT', 'ROU', 'SVK', 'SVN', 'SWE']
}

earliest = pd.to_datetime('1998-01-01')
new_cols = ['country_name', 'election_date', 'party_id', 'party_name_short', 'vote_share']

df = pd.read_csv(url, parse_dates=['election_date'])

for col, values in selection.items():
    df = df[df[col].isin(values)]

df = df[df['election_date'] >= earliest]
df = df[new_cols]
df = df.dropna().reset_index(drop=True)

df.head()
```

## 2. Add populism classification

The data comes from [The PopuList](https://popu-list.org/), a project initiated by [The Guardian](https://www.theguardian.com/world/ng-interactive/2018/nov/20/revealed-one-in-four-europeans-vote-populist) and supported by the [Amsterdam Institute for Social Science Research](https://aissr.uva.nl/) and the [ECPR Standing Group on Extremism and Democracy](https://ecpr.eu/StandingGroups/StandingGroupHome.aspx?ID=6). Their list consists of European parties that can be classified as populist, far right, far left and/or Eurosceptic, and obtained at least 2% of the vote in at least one national parliamentary election since 1998.

```python
df_pop = pd.read_excel('../input/populist-version-1-20190211.xlsx', dtype={'Populist': bool, 'Far right': bool, 'Far left': bool, 'Eurosceptic': bool})
df_pop.head()
```

Spains new right-wing party Vox isn't included in the PopuList dataset, yet. We append it manually:

```python
df_pop = df_pop.append({'Country': 'Spain', 'Party': 'Vox', 'Populist': True, 'Far right': True, 'Far left': False, 'Eurosceptic': True}, ignore_index=True)
```

Unfortunatelly the partys' names in the ParlGov and the PopuList dataset don't match up in all cases. The Guardian provided us with [a lookup table](../input/parlogv_populist_lookuptable.xlsx) to merge them via the partys' ParlGov ID. (__Thank you for that!__) We refined and updated some information by hand and added the comments column from the [PopuList's PDF file](https://populistorg.files.wordpress.com/2019/02/populist-version-1-20190211.pdf) that says if a classification applies to a certain period of time, only. As a result we got [this clean lookup table for parties in the EU28](../input/parlogv_populist_lookup.csv):

```python
df_lookup = pd.read_csv('../input/parlogv_populist_lookup.csv')
df_lookup.head()
```

So, let's join the datasets! First we add a new column “Party” to our results dataframe and fill it with a party's short name from the PopuList dataset if it appears in it. If the “comments” column says the classification applies to a certain period of time, we only fill in the new column for the elections in this certain period of time.

```python
def get_join_name(row):
    df_results = df_lookup[df_lookup['party_id'] == row['party_id']]
    if len(df_results) == 0:
        return np.nan
    if len(df_results) > 1:
        return 'ERROR'
    result = df_results.iloc[0]
    if pd.isnull(result['comments']):
        return result['party']
    comment = result['comments'].split()
    if comment[0] == 'until':
        if row['election_date'].year < int(comment[1]):
            return result['party']
        return np.nan
    if comment[0] == 'in':
        if row['election_date'].year == int(comment[1]):
            return result['party']
        return np.nan
    if comment[0] == 'since':
        if row['election_date'].year >= int(comment[1]):
            return result['party']
        return np.nan
    return 'ERROR'

df['Party'] = df.apply(get_join_name, axis=1)
```

Any errors?

```python
df[df['Party'] == 'ERROR']
```

Good. Now we merge the prepared election results with the PopuList's classification and fill up the not classified partys' cells with `False`s:

```python
df = df.merge(df_pop, how='left', left_on=['country_name', 'Party'], right_on=['Country', 'Party'])
df = df[['country_name', 'election_date', 'party_name_short', 'vote_share', 'Populist', 'Far right', 'Far left', 'Eurosceptic']]
df = df.sort_values('vote_share', ascending=False).sort_values(['country_name', 'election_date'])

for col in ['Populist', 'Far right', 'Far left', 'Eurosceptic']:
    df.loc[df[col].isnull(), col] = False

df.to_csv('../output/election_results.csv', index=False, float_format='%.2f')
```

## 3. Analyses

### 3.1 Small Multiples

```python
colors = {
    '1_pop_right': '#80b1d3',
    '2_pop_center': '#fdb462',
    '3_pop_left': '#fb8072',
    '4_eurosceptic': '#aaaaaa',
    '5_other': '#e2e2e2'
}

def get_color(row):
    c.append(colors[row['type']])

fig = plt.figure(figsize=(20, 30))
ax = {}
i = 1
cols = 4
rows = int(np.ceil(len(df['country_name'].value_counts().sort_index().index) / cols))

for country in df['country_name'].value_counts().sort_index().index:
    df_c = df[df['country_name'] == country]
    df_c['type'] = '5_other'
    df_c.loc[df_c['Eurosceptic'], 'type'] = '4_eurosceptic'
    df_c.loc[df_c['Populist'], 'type'] = '2_pop_center'
    df_c.loc[df_c['Far right'], 'type'] = '1_pop_right'
    df_c.loc[df_c['Far left'], 'type'] = '3_pop_left'
    df_c['normal_dup'] = df_c['election_date']
    df_c.loc[df_c['type'] != '5_other', 'normal_dup'] = np.nan
    df_c = df_c[(~df_c[['normal_dup']].duplicated()) | (df_c['normal_dup'].isnull())]
    df_c = df_c[['election_date', 'type', 'vote_share']]
    for election in df_c['election_date'].value_counts().sort_index().index:
        vote_sum = df_c[(df_c['election_date'] == election) & (df_c['type'] != '5_other')]['vote_share'].sum()
        df_c.loc[(df_c['election_date'] == election) & (df_c['type'] == '5_other'), 'vote_share'] = 100 - vote_sum
    df_c = df_c.sort_values('vote_share', ascending=False).sort_values(['election_date', 'type']).reset_index(drop=True).reset_index()
    c = []
    df_c.apply(get_color, axis=1)
    df_c = df_c.pivot(values='vote_share', index='election_date', columns='index')
    ax[country] = fig.add_subplot(rows, cols, i)
    i += 1
    df_c.plot(kind='bar', stacked=True, title=country, width=1, color=c, linewidth=1, edgecolor='#ffffff', legend=False, ax=ax[country])
    ax[country].tick_params(
        which='both',
        bottom='off',
        left='off',
        right='off',
        top='off'
    )
    ax[country].set_ylabel('')
    ax[country].set_xlabel('')
    ax[country].spines['left'].set_visible(False)
    ax[country].spines['top'].set_visible(False)
    ax[country].spines['right'].set_visible(False)
    ax[country].xaxis.set_major_formatter(ticker.FixedFormatter([item.strftime('%Y') for item in df_c.index]))

plt.savefig('../output/small_multiples.pdf')
```

### 3.2 Heatmap

```python
df_steps = df[df['Populist']].groupby(['country_name', 'election_date'])['vote_share'].sum().reset_index()
df_steps['value'] = 0
for i in range(0, 6):
    df_steps.loc[df_steps['vote_share'] > i * 10, 'value'] = i + 1
df_steps = df_steps[['country_name', 'election_date', 'value']]

df_popres = df.groupby(['country_name', 'election_date'])['vote_share'].count().reset_index()[['country_name', 'election_date']]
df_popres = df_popres.merge(df_steps, how='left', on=['country_name', 'election_date'])
df_popres = df_popres.fillna(0)
df_popres['election_date'] = df_popres['election_date'].astype(str).str[:-3]
df_popres = df_popres.pivot(values='value', columns='country_name', index='election_date').reset_index()

df_popres.head(2)
```

```python
months = []

for year in range(1998, 2020):
    for month in range(1, 13):
        months.append(str(year) + '-' + f'{month:02d}')

df_months = pd.DataFrame(months)
df_months.columns = ['election_date']

df_heat = df_months.merge(df_popres, how='left', on='election_date').set_index('election_date')
df_heat = df_heat.ffill()

df_heat.head(4)
```

Show the populist parties' sums of the voting shares for each country over time:

```python
fig = plt.figure(figsize=(20, 10))
sns.heatmap(df_heat[2:-6].transpose().fillna(7), cmap=['#ffffff', '#eddef5', '#dbbdeb', '#c89ce1', '#b47bd7', '#9f5acc', '#8837c1', '#eeeeee'])

plt.savefig('../output/heatmap.pdf')
```
